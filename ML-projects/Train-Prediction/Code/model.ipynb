```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
df = pd.read_csv('../Source/Travel.csv', header=0)
df
```

```python
df['Gender'] = df['Gender'].replace('Fe Male', 'Female')
df['MaritalStatus'] = df['MaritalStatus'].replace('Single', 'Unmarried')
```

```python
# Finding the features with null entries
null_features = [ features for features in df.columns if df[features].isnull().sum() >=1 ]

for feature in null_features:
  print(feature, np.round(df[feature].isnull().mean() * 100, 5), '% missing null values')
```

```python
# Getting all the details of the null features
df[null_features].select_dtypes(exclude='object').describe()
```

```python
# For median imputation
df['Age'] = df['Age'].fillna(df['Age'].median())
df['DurationOfPitch'] = df['DurationOfPitch'].fillna(df['DurationOfPitch'].median())
df['NumberOfTrips'] = df['NumberOfTrips'].fillna(df['NumberOfTrips'].median())
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())

# For mode imputation
df['NumberOfFollowups'] = df['NumberOfFollowups'].fillna(df['NumberOfFollowups'].mode()[0])
df['NumberOfChildrenVisiting'] = df['NumberOfChildrenVisiting'].fillna(df['NumberOfChildrenVisiting'].mode()[0])
df['PreferredPropertyStar'] = df['PreferredPropertyStar'].fillna(df['PreferredPropertyStar'].mode()[0])
```

```python
df['TypeofContact'] = df['TypeofContact'].fillna(df['TypeofContact'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.drop('CustomerID', axis=1, inplace=True)
```

# Feature Engineering

```python
# We are reducing the number of columns by merging the number of children visited and number of persons visited into one column
df['TotalVisiting'] = df['NumberOfChildrenVisiting'] + df['NumberOfPersonVisiting']
df.drop(columns=['NumberOfPersonVisiting', 'NumberOfChildrenVisiting'], axis=1, inplace=True)
```

```python
# Number of numerical features
num_features = [ features for features in df.columns if df[features].dtype != 'O']
print(len(num_features))
print(num_features)
```

```python
# Number of categorical features
num_features = [ features for features in df.columns if df[features].dtype == 'O']
print(len(num_features))
print(num_features)
```

```python
# Discrete Features - the ones that have atleast about 10 or 25 different values
num_features = [ features for features in df.columns if len(df[features].unique()) <= 25]
print(len(num_features))
print(num_features)
```

# Model

```python
from sklearn.model_selection import train_test_split
x = df.drop('ProdTaken', axis=1)
y = df['ProdTaken']
```

```python
y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
x_train
```

```python
# Encoding
cat_features = x.select_dtypes(include='object').columns
num_features = x.select_dtypes(exclude='object').columns
```

```python
# Standaradisation
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

oh_encoder = OneHotEncoder(drop='first')
numeric_encoder = StandardScaler()

preprocessor = ColumnTransformer(
  [
    ('OneHotEncoder', oh_encoder, cat_features),
    ('StandardScaler', numeric_encoder, num_features)
  ]
)
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score, f1_score, precision_score, recall_score
```

```python
models = {
  'Logistic Regression': LogisticRegression(),
  'Decision Tree': DecisionTreeClassifier(),
  'Random Forest': RandomForestClassifier()
}

for model_name, model in models.items():
  # Fit the model
  model.fit(x_train, y_train)
  model.fit(x_train, y_train)

  y_pred_train = model.predict(x_train)
  y_pred_test = model.predict(x_test)

  print(model_name)
  # Trainig data
  train_accuracy = accuracy_score(y_train, y_pred_train)
  print(f"Accuracy: {train_accuracy:.4f}")
  train_precision = precision_score(y_train, y_pred_train, average='weighted', zero_division=0)
  train_recall = recall_score(y_train, y_pred_train, average='weighted', zero_division=0)
  train_f1 = f1_score(y_train, y_pred_train, average='weighted', zero_division=0)
  print(f"Precision (weighted): {train_precision:.4f}")
  print(f"Recall (weighted): {train_recall:.4f}")
  print(f"F1-Score (weighted): {train_f1:.4f}")

  # Confusion Matrix
  train_conf_matrix = confusion_matrix(y_train, y_pred_train)
  print("\nConfusion Matrix:\n", train_conf_matrix)

  # Test data

  test_accuracy = accuracy_score(y_test, y_pred_test)
  print(f"Accuracy: {test_accuracy:.4f}")

  # Precision, Recall, F1-Score
  test_precision = precision_score(y_test, y_pred_test, average='weighted', zero_division=0)
  test_recall = recall_score(y_test, y_pred_test, average='weighted', zero_division=0)
  test_f1 = f1_score(y_test, y_pred_test, average='weighted', zero_division=0)
  print(f"Precision (weighted): {test_precision:.4f}")
  print(f"Recall (weighted): {test_recall:.4f}")
  print(f"F1-Score (weighted): {test_f1:.4f}")

  # Confusion Matrix
  test_conf_matrix = confusion_matrix(y_test, y_pred_test)
  print("\nConfusion Matrix:\n", test_conf_matrix)

  # Classification Report
  print("\nClassification Report:\n", classification_report(y_test, y_pred_test, zero_division=0))

  print('='*35)
  print('\n')
```

```python
rf_params = {
  'criterion': ["gini", "entropy", "log_loss"],
  'n_estimators': [50, 100, 150, 200, 250, 300, 400, 500, 1000],
  'max_depth': [5, 8, 10, 15, 20, None],
  'ccp_alpha': [1, 2, 3, 4, 6, 7, 8, 10, 12, 0],
  'max_features': ["sqrt", "log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
randomcv_params = [
  ("RF", RandomForestClassifier(), rf_params),
]
```

```python
model_params = {}
for name, model, params in randomcv_params:
  grid = GridSearchCV(estimator=model, param_grid=params, cv=5, verbose=2, n_jobs=-1, refit=True)

  grid.fit(x_train, y_train)
  model_params[name] = grid.best_params_

for model_name in model_params:
  print(f"------------------ Best parameters for {model_name} --------------------")
  print(model_params[model_name])
```
