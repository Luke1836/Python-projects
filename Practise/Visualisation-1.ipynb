```python
# Creating a line plot
x = [1,2,3,4,5]
y = [3,4,52,3,5]
plt.plot(x, y, color='purple', linestyle='-', marker='o', linewidth=2, markersize=9)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic line plot")
```

```python
## Creating multiple plots
x = [1,2,3,4,5,6,7,8]
y1 = [12,10,16,8,21,23,25,14]
y2 = [2,5,9,21,11,23,10,18]

plt.figure(figsize=(10,3))

plt.subplot(1,2,1)  # One row and 2 column
plt.plot(x,y1,color='blue', marker="o", markerfacecolor='red')
plt.yticks([10,15,20,25])
plt.title('Plot-1')

plt.subplot(1,2,2)  # One row and 2 column
plt.plot(x,y2,color='blue', marker="o", markerfacecolor='red')
plt.yticks([0,5,10,15,20,25])
plt.title('Plot-2')
```
